Importing the libraries


```python
import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from wordcloud import wordcloud
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

Loading the dataset

```python
# Load both real and fake news datasets
fake_df = pd.read_csv("Fake.csv")
true_df = pd.read_csv("True.csv")

# Add labels: 0 = Fake, 1 = Real
fake_df["label"] = 0
true_df["label"] = 1

# Combine the datasets
data = pd.concat([fake_df, true_df], ignore_index=True)
data = data.sample(frac=1).reset_index(drop=True)  # Shuffle the data

data.head()
```

Cleaning the Text

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\n', ' ', text)
    return text

# Apply cleaning
data['cleaned_text'] = data['text'].apply(clean_text)
```

Preprocessing the data using NLP

```python
import re
nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = stopwords.words("english")


def clean_text(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)      # remove URLs
    text = re.sub(r'<.*?>', '', text)                      # remove HTML tags
    text = re.sub(r'[^a-z\s]', '', text)                   # remove special characters & digits
    text = re.sub(r'\s+', ' ', text).strip()               # remove extra whitespace
    return text

# Apply it to your dataset
data['cleaned_text'] = data['text'].apply(clean_text)
```

split dataset

```python
X = data['cleaned_text']
y = data['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
print("Train shape:",X_train.shape)
print("Test shape:",X_test.shape)
```

TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
print(X.shape)
print(y.value_counts())
```

Training the Machine Learning Model

```python
model = LogisticRegression()
model.fit(X_train_vec, y_train)
```

Evaluating the model

```python
y_pred = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['FAKE', 'REAL'], yticklabels=['FAKE', 'REAL'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Confusion Matrix")
plt.show()
```

Predicting the model

```python
def predict_news(text):
    cleaned = clean_text(text)
    vect = vectorizer.transform([cleaned])
    prediction = model.predict(vect)[0]

    if prediction == 0:
        return "FAKE"
    else:
        return "REAL"

# Test
sample = "The Prime Minister announced a new policy for education today."
print("Prediction:", predict_news(sample))
```

```python
def predict_news(text):
    # Clean the text using the same preprocessing function
    cleaned = clean_text(text)
    # Convert to vector using the fitted TF-IDF vectorizer
    vector = vectorizer.transform([cleaned]).toarray()
    # Predict
    prediction = model.predict(vector)
    return "REAL" if prediction[0] == 1 else "FAKE"
```
